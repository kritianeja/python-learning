```python
#products = [10, 25, 0, 12, 9]   # 0 means out of stock
products=[10, 25, 0, 12, 9] 
for i in products:
    if i==0:
        break
    print(i)
```

```python
#temps = [78, 85, 92, 101, 95]
temps=[78, 85, 92, 101, 95]
for i in temps:
    if i>100:
        break
    print(i)
```

```python
#transactions = [500, 900, 1200, -300, 800]
transactions = [500, 900, 1200, -300, 800]
for i in transactions:
    if i<0:
        break
    print(i)
```

```python
# Print only valid email addresses (skip emails without '@').
emails = ["abc@gmail.com", "test", "user@xyz.com", "invalid"]
for i in emails:
    if "@" not in i or ".com" not in i:
        continue
    print(i)
```

```python
#Print only even numbers from 1 to 10 (skip odd numbers).
for i in range(1,11):
    if i%2!=0:
        continue
    print(i)
```

```python
#Skip customers who have zero purchase amount. Print only paying customers.

amounts = [200, 0, 450, 0, 300]
for i in amounts:
    if i==0:
        continue
    print(i)
```

```python
#Skip all empty strings while printing customer names.
names = ["Ravi", "", "Suman", "", "Kiran"]
for i in names:
    if i=="":
        continue
    print(i)
```

```python
#Print the only first number which is divisible by both 5 and 9 between 1 - 100
for i in range(1,101):
    if i%5==0 and i%9==0:
        print(i)
        continue
        
        
```
